```python
Q1 = placement["IQ"].quantile(0.25)
Q3 = placement["IQ"].quantile(0.75)
IQR = Q3 - Q1
print("The lower quartile of Age is: ", Q1)
print("The upper quartile of Age is:", Q3)
print('The interquartile range of age is:', IQR)
upper_bound = Q3 + 1.5*IQR
lower_bound = Q1 - 1.5*IQR
print(f"Lower bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
```

```python
no_outliers = placement[~(placement['IQ']>=upper_bound)]
no_outliers
```

```python
plt.boxplot(no_outliers['IQ'])
```
