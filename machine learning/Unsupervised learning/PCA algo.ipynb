# 08/11/23 - Principal Component Analysis
Principal Components analysis or PCA is a popular dimensionality reduction technique you can use to avoid "curse of dimensionality".<br>
The curse of dimensionality occurs when we have too many feature and less data points.<br>
It is a data preprocessing algo<br>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv("../../dataset/creditcard.csv")
df.head()
```

```python
X = df.drop(columns=['Class'])
X.head()
```

```python
ss = StandardScaler()
new_data = ss.fit_transform(X)
new_data
```

```python
pca = PCA(n_components=3)#components
pcs = pca.fit_transform(X)
pcs
```

```python
principalDF = pd.DataFrame(pcs,columns=['PC1','PC2','PC3'])
principalDF.head()
```

```python
pca.explained_variance_
```

```python
pca_model = PCA(n_components=0.9)# variance
pca_model.fit_transform(new_data)
pca_model
```

```python
pca_model.explained_variance_
```
