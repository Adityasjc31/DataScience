## Method2: Rename a column

```python
df3 = df.copy()
df3.head()
```

```python
df3.rename({"Administration":"Admin","State":"MyState"},
           axis=1,
           inplace=True)
df3.head()
```

```python
df3.rename(columns={"Admin":"Administration","MyState":"State"},
           inplace=True)
df3.head()
```
