```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from wordcloud import WordCloud,STOPWORDS,ImageColorGenerator
%matplotlib inline
```

```python
instagram = pd.read_csv('../dataset/Instagram data.csv',encoding="latin1")
instagram.head(1)
```

```python
instagram.columns
```

```python
instagram.info()
```

```python
instagram.isna().sum()
```

```python
plt.figure(figsize=(7,5))
plt.style.use("seaborn-v0_8-muted")
sns.distplot(instagram['From Home'])
plt.title("Distribution of Impressions from home")
```

```python
plt.figure(figsize=(7,5))
plt.style.use("seaborn-v0_8-muted")
sns.distplot(instagram['From Explore'])
plt.title("Distribution of Impressions from explore")
```

```python
txt = " ".join(i for i in instagram['Caption'])
txt
```

```python
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords,background_color="white").generate(txt)
plt.style.use("classic")
plt.figure(figsize=(15,12))
plt.imshow(wordcloud,interpolation="bilinear")
plt.axis("off")
```
